```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder


class Modify(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass
    
    def fit(self, X, y=None):
        return self
    
    
    def transform(self, X, y=None):
        X['Sex'] = X['Sex'].map({'female': 0, 'male': 1})
#         아래는 왜 안되는지 모르겠음
#         X[X['Sex'] == 'male']['Sex'] = 1
#         X[X['Sex'] != 'male']['Sex'] = 0
        X['family'] = X['SibSp'] + X['Parch']
        print(X)
        X.drop(['SibSp', 'Parch'], axis=1, inplace=True)
        print(X)
        return X

num_attribs = ['Pclass', 'Sex', 'Fare', 'SibSp', 'Age', 'Parch']
cat_attribs = ['Embarked']
# 나머지는 drop

num_pipeline = Pipeline([
    ('modify_sex', Modify()),
    ('imputer', SimpleImputer(strategy='mean')),
])

full_pipeline = ColumnTransformer([
    ('num_pipline', num_pipeline, num_attribs),
    ('cat', OneHotEncoder(categories='auto'), cat_attribs),
])
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics


X_train = full_pipeline.fit_transform(df_train)  # fit_transform()은 학습데이터로
label_train = df_train['Survived'].values 

X_tr, X_vld, y_tr, y_vld = train_test_split(X_train, label_train, test_size=0.2, random_state=42)

from sklearn.model_selection import GridSearchCV


param_grid = [
    {'n_estimators': [3, 10, 30], 'max_features': [2, 4, 6, 8]},
    {'bootstrap': [False], 'n_estimators': [10, 30], 'max_features': [4, 5, 6, 7]},
  ]

forest_cls = RandomForestClassifier(random_state=42)
grid_search = GridSearchCV(forest_cls, param_grid, cv=5, scoring='accuracy', 
                           return_train_score=True, n_jobs=-1)
grid_search.fit(X_tr, y_tr)
```

```python
cvres = grid_search.cv_results_
for score, params in zip(cvres["mean_train_score"], cvres["params"]):  # zip([a,b,c], [1,2,3]) => (a, 1), (b, 2), (c, 3) 반환됨
    print(score, params)
print(grid_search.best_params_)
```

```python
model = grid_search.best_estimator_
model.fit(X_tr, y_tr)
prediction = model.predict(X_vld)
```

```python
print('총 {}명 중 {:.2f}% 정확도로 생존을 맞춤'.format(y_vld.shape[0], 100 * metrics.accuracy_score(prediction, y_vld)))
```
